### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "../Resources/city_data.csv"
ride_data_to_load = "../Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_ride = pyber_data_df.groupby(["type"]).sum()["fare"] / pyber_data_df.groupby(["type"]).count()["ride_id"]
average_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_driver = pyber_data_df.groupby(["type"]).sum()["fare"] / city_data_df.groupby(["type"]).sum()["driver_count"]
average_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_totals = {'Total Rides': total_rides, 'Total Drivers': total_drivers, 'Total Fares': total_fares, 'Average Fare by Ride': average_fare_ride, 'Average Fare by Driver': average_fare_driver}
pyber_summary_df = pd.DataFrame(pyber_totals)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df.dtypes
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map('${:,.2f}'.format)
pyber_summary_df["Average Fare by Ride"] = pyber_summary_df["Average Fare by Ride"].map('${:,.2f}'.format)
pyber_summary_df["Average Fare by Driver"] = pyber_summary_df["Average Fare by Driver"].map('${:,.2f}'.format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# Instructions don't match step-by-step instructions in challenge. Step 1
# says to create a new DataFrame with multiple indices using groupby() function
# on the "type" and "date" columns of the pyber_data_df then apply sum() method
# on the "fare" column to show the total fare amount for each date.
date_fare_df = pd.DataFrame({'Sum of Fares': pyber_data_df.groupby(["type","date"]).sum()["fare"]})
date_fare_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# Step 2 step by step instructions say to reset the index with provided code
# that is actually listed in step 3.
date_fare_df = date_fare_df.reset_index()
date_fare_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
# Step 3 in step by step indicates to use the pivot() function to create the new DataFrame
date_fare_pivot = date_fare_df.pivot(index="date", columns="type", values="Sum of Fares")
date_fare_pivot
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
# Step by step says to create a new dataframe by using loc method on date range
# 2019-01-01 through 2019-04-29.
start_date = "2019-01-01"
end_date = "2019-04-29"
#date_fare_pivot_filter = date_fare_pivot[date_fare_pivot.date >= start_date]

after_start_date = date_fare_pivot.index >= start_date
before_end_date = date_fare_pivot.index <= end_date
between_two_dates = after_start_date & before_end_date
filtered_dates = date_fare_pivot.loc[between_two_dates]
filtered_dates
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
# Step by step says to use the provided code to reset the index from the
# previous step using the provided code to a datetime data type.
filtered_dates.index = pd.to_datetime(filtered_dates.index)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
# Step by step says to use df.info() to check the date is a datetime data type.
filtered_dates.info()
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
# Step by step says to create a new DataFrame by applying the resample()
# function to the modified DataFrame from step 5. Resampling to weekly bins
# then apply the sum() method to get the total fares for each week.
filtered_dates_weekly = filtered_dates.resample("W").sum()
filtered_dates_weekly
```
